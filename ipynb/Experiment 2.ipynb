#### c) Draw a scatter plot for given data that shows the number of games played and scores obtained. Write outlier data.

```python
from matplotlib import pyplot as pt
x=[5,8,1,9,7,4,8,6,7]
y=[90,87,56,88,45,79,90,79,80]
pt.scatter(x,y,marker=".",s=150,c="k")
pt.title("Scatter plot for number of games played and scores obtained")
pt.xlabel("Number of Games")
pt.ylabel("Score")
pt.grid()
pt.show()
```

### Outlier data in above scatter plot is
##### Points (1,56) and (7,45)

#### d) Draw a scatter plot for total bills against tips. Use tips dataset. 

```python
import pandas as pd
df=pd.read_csv('tips.csv')
df.head()
```

```python
from matplotlib import pyplot as pt
x=df['total_bill']
y=df['tip']
pt.scatter(x,y,marker=".",s=100,c="k")
pt.title("Scatter Plot for total_bill against tip")
pt.xlabel("Total Bill")
pt.ylabel("Tips")
pt.grid()
pt.show()
```
